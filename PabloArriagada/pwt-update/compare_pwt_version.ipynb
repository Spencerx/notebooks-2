```python
import pandas as pd

pwt_old = pd.read_excel("pwt100.xlsx", sheet_name='Data')
pwt_new = pd.read_excel("pwt1001.xlsx", sheet_name='Data')
```

```python
pwt_new
```

```python
# Compare the two dataframes and show the differences
pwt_diff = pwt_old.compare(pwt_new)
pwt_diff
```

```python
for col in pwt_diff.columns:
    print(col)
```

```python
# Get number of non-null values for rtfpna and years

print(f"10.0: years [{pwt_old.year.min()}, {pwt_old.year.max()}], count: {pwt_old.year.count()}")
print(f"10.01: years [{pwt_new.year.min()}, {pwt_new.year.max()}], count: {pwt_new.year.count()}")
```

```python
# In plotly, plot a scatter plot of the variable rtfpna for both dataframes
pwt_plot = pd.merge(pwt_old[['countrycode', 'year', 'rtfpna']], pwt_new[['countrycode', 'year', 'rtfpna']], on=['countrycode', 'year'], suffixes=('_old', '_new'))

import plotly.express as px
fig = px.scatter(pwt_plot, x='rtfpna_old', y='rtfpna_new', hover_data=['countrycode', 'year'], color='countrycode')
# Add a 45 degree line
fig.add_shape(type='line', x0=0, y0=0, x1=8, y1=8, line=dict(color='black', width=1), row='all', col='all')
fig.update_layout(title='Scatter plot of `rtfpna` for PWT 10.0 and PWT 10.1')
fig.show()
```

```python
# Check how many values are different
pwt_plot['diff'] = pwt_plot['rtfpna_new'] == pwt_plot['rtfpna_old']

pwt_plot['diff'].dropna().value_counts()
```

```python
# Plot line charts for the variable rtfpna for both dataframes

fig = px.line(pwt_old, x='year', y='rtfpna', hover_data=['countrycode'], color='countrycode')
fig.show()

fig = px.line(pwt_new, x='year', y='rtfpna', hover_data=['countrycode'], color='countrycode')
fig.show()
```
