```python
import pandas as pd

# Define questions
QUESTIONS = ['CONTENT', 'HAPPY', 'LIFE_SAT', 'MENTAL_HEALTH', 'PHYSICAL_HLTH', 'WORRY_SAFETY', 'EXPENSES']

# Define dictionary of countries
COUNTRIES_AND_CODES = {
    1: "Argentina",
    2: "Australia",
    3: "Brazil",
    4: "Egypt",
    5: "Germany",
    6: "India",
    7: "Indonesia",
    8: "Israel",
    9: "Japan",
    10: "Kenya",
    11: "Mexico",
    12: "Nigeria",
    13: "Philippines",
    14: "Poland",
    16: "South Africa",
    17: "Spain",
    18: "Tanzania",
    19: "Turkey",
    20: "United Kingdom",
    22: "United States",
    23: "Sweden",
    24: "Hong Kong"
}

# Open csv files

df_collapse = pd.read_csv("/Users/parriagadap/notebooks/PabloArriagada/global_flourishing_study/gfs_collapse.csv")
df_svy = pd.read_csv("/Users/parriagadap/notebooks/PabloArriagada/global_flourishing_study/gfs_svy_original.csv")

# Edit df_svy
# Separate survey column by the "-" character. Name columns COUNTRY AND WAVE
df_svy[['COUNTRY', 'WAVE']] = df_svy['survey'].str.split('-', expand=True)
df_svy = df_svy.drop(columns=['survey'])

# Change country codes to country names
df_svy['COUNTRY'] = df_svy['COUNTRY'].astype(int)
df_svy['COUNTRY'] = df_svy['COUNTRY'].map(COUNTRIES_AND_CODES)

# Make WAVE in the format of Year {number}
df_svy['WAVE'] = "Year " + df_svy['WAVE']

# Merge df_collapse and df_svy
df = pd.merge(df_collapse, df_svy, on=['COUNTRY', 'WAVE'], suffixes=('_collapse', '_svy'))

# For each question, calculate the difference between the collapsed and survey data
for question in QUESTIONS:
    df[f'{question}_diff'] = df[f'{question}_collapse'] - df[f'{question}_svy']
    df[f'{question}_ratio'] = df[f'{question}_collapse'] / df[f'{question}_svy']

df.to_csv("/Users/parriagadap/notebooks/PabloArriagada/global_flourishing_study/gfs_diff.csv", index=False)
```
